```python
np.random.seed(42)
N = [10,50,100,500,1000,5000,10000,50000,100000]
def toss_coin(n):
  event = np.random.randint(2,size=n)
  trial = np.arange(1,n+1)
  relative = np.cumsum(event)/trial
  prob = np.sum(relative)/n
  print(f"n = {n}, p(e) = {np.around(prob,3)}")
  if n == 1000:
    plt.plot(relative)
    plt.grid()
    plt.axhline(y=0.5,color='r')
  return prob,trial,relative

for n in N:
  prob,trial,relative = toss_coin(n)
  print("-----------------------------------------")
  
  #plt.plot(trial, prob, color='blue', marker='o')

```

2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
def fair_die(n,number):
  event = np.random.randint(6,size=n)+1
  trial = np.arange(1,n+1)
  prob = np.count_nonzero(event == 5)/n
  relative = np.cumsum(event == 5)/trial
  print(f"n = {n}, probability of number {number} is {prob}")
  if n == 1000:
    plt.plot(relative)
    plt.grid()
    plt.axhline(y=prob,color='r')

for n in N:
  fair_die(n,5)
  print("-----------------------------------------")
```

# Problem 3: Independence

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python
#type your code here:

# simulate 1 million tests of two fair coin flips
n = 1000000
event = np.random.randint(2,size=(n,2))
# sums of all tests
test = (event.sum(axis=1) == 2).sum()
# proportion of tests that produced exactly two heads
test.mean()/n
```

2. Three fair coin flips produce exactly one head

```python
#type your code here:

# simulate 1 million tests of three fair coin flips
n = 1000000
event = np.random.randint(2,size=(n,3))

# sums of all tests
test = (event.sum(axis=1) == 1).sum()
# proportion of tests that produced exactly one head
test.mean()/n
```

# Problem 5: Correlation and Covariance

The CSV file includes a hourly/daily summary for Szeged, Hungary area, between 2006 and 2016. After inspecting the data:

- Calculate the covariance and correlation coeffecient between temperature and apparent temperature
- Calculate the covariance and correlation coeffecient between temperature and humidity

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
apparant_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
```

```python
plt.scatter(temperature, apparant_temp)
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
```

```python
plt.scatter(humidity, temperature)
plt.grid()
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
```

## Temperature and Apparent Temperature

```python
#type your code here:
m = temperature.size
a = temperature
b = apparant_temp

cov = ((a.T - a.mean())@(b - b.mean())) / (m-1)
print(cov)
print(np.cov(a.T,b.T))

corr = cov / (np.std(a.T) * (np.std(b.T)))
print(corr)
print(np.corrcoef(a.T,b.T))
```
